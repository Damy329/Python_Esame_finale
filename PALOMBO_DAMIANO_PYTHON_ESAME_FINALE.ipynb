### 2.
  #### Si chiede di trovare, per ogni continente:
   * a. il numero di casi fin dall'inizio della pandemia
   * b. la percentuale rispetto al totale mondiale del numero di casi

```python
# a) Numeri casi totali per continente
numero_casi_totali_per_continente= df.groupby('continent')['new_cases'].sum()
print(f'Il numero di casi totali per continente è: \n{numero_casi_totali_per_continente}')
```

```python
# Numeri casi totali mondiali
numero_casi_totali_mondiale= df['new_cases'].sum()
print(f'Il numero di casi totali mondiali è: {numero_casi_totali_mondiale}')
```

```python
# b) Percentuale del numero dei casi rispetto al totale mondiale 
percentuale_statistica= (numero_casi_totali_per_continente / numero_casi_totali_mondiale)*100
print(percentuale_statistica)
print('\n') # aggiungo uno spazio tra i due output per maggiore leggibilità
# Formatto l'output come percentuale con due decimali
print(f'La percentuale di ogni continente rispetto al totale mondiale è: \n')
for continente, percentuale in percentuale_statistica.items():
    print(f'{continente}: {percentuale:.2f}%') 
    # Per l'output scelgo di visualizzare le due variabili in un dizionario, arrotondando per eccesso a due cifre decimali seguite dal simbolo %

```

### 3. 
#### Selezionare i dati relativi all'Italia nel 2022 e mostrare con dei grafici adeguati:
   * a. l'evoluzione del casi totali dall'inizio alla fine dell'anno
   * b. il numero di nuovi casi rispetto alla data; poiché i nuovi casi vengono registrati settimanalmente, filtrare via i dati che non hanno misurazioni

```python
import matplotlib.pyplot as plt
# Converto la colonna 'date' in formato datetime
df['date'] = pd.to_datetime(df['date'])

# Filtro i dati per l'Italia e l'anno 2022
andamento_covid_Italia_2022 = df[(df['location'] == 'Italy') & (df['date'].dt.year == 2022)]

# Visualizzazione grafica dell' evoluzione dei casi totali dall'inizio alla fine del 2022 in Italia
plt.figure(figsize=(10,6))
plt.plot(andamento_covid_Italia_2022['date'], andamento_covid_Italia_2022['total_cases'], label='Casi Totali', color='green') 
plt.title('Evoluzione Casi Totali in Italia nel 2022')
plt.xlabel('Data')        # etichette asse x
plt.ylabel('Casi Totali') # etichette asse y 
plt.xticks(rotation=45)   # rotazione delle etichette sull'asse delle x
plt.grid(True)            # visualizzazione della griglia
plt.tight_layout()        # layout
plt.legend()              # visualizzazione leggenda
plt.show()

#altro approccio di visualizzazione equivalente con seaborn

print('Altro approccio equivalente con seaborn: \n')

import seaborn as sns
sns.lineplot(data=andamento_covid_Italia_2022, x='date', y='total_cases')

# filtro per visualizzare l'Italia e dati  non nulli o diversi da 0
df_italia_2022_nuovo = andamento_covid_Italia_2022[(andamento_covid_Italia_2022['new_cases'].notnull())
& (andamento_covid_Italia_2022['new_cases'] !=0)]


# Visualizzazione grafica dei nuovi casi filtrati in Italia nel 2022
plt.figure(figsize=(10,6))
plt.plot(df_italia_2022_nuovo['date'], df_italia_2022_nuovo['new_cases'], label='Nuovi Casi', color='red', linewidth=3) 
# paramentro: linewidth, aumenta lo spessore della riga
#settato un colore definito a piacimento
plt.title('Nuovi Casi in Italia nel 2022')
plt.xlabel('Data')           # nome asse x
plt.ylabel('Nuovi Casi')     # nome asse y 
plt.xticks(rotation=45)      # rotazione etichette sull' asse x 
plt.grid(True)               # visualizzazione della griglia
plt.tight_layout()           # layout
plt.legend()                 # visualizzazione leggenda
plt.show()                   # inserisco questo metodo per avere un output più pulito senza lista di valori che sono assegnati agli assi


#altro approccio di visualizzazione equivalente con seaborn

sns.lineplot(data=df_italia_2022_nuovo, x='date', y='new_cases')



```

### Analisi Aggiuntive extra

### Analisi tasso di vaccinazione
Per ogni continente:
 * Confronta la percentuale di popolazione completamente vaccinata.

```python
# Filtro i dati per togliere valori nulli per ogni continente
df_latest = df.dropna(subset=['continent', 'people_fully_vaccinated_per_hundred', 'total_deaths', 'total_cases'])

# Oridnamento crescente per continente 
df_latest = df_latest.sort_values('date').groupby('continent').last()

# Calcolo della percentuale di popolazione completamente vaccinata per continente
vaccination_rate = df_latest['people_fully_vaccinated_per_hundred']

# Visualizzazione dei dati di vaccinazione
plt.figure(figsize=(10, 6))
vaccination_rate.sort_values().plot(kind='bar', color='#0000FF')   # paramentro: color, settato il blu in esadecimali
plt.title('Percentuale di popolazione completamente vaccinata per continente')
plt.xlabel('Continente')
plt.ylabel('Percentuale di popolazione completamente vaccinata (%)')
plt.xticks(rotation=45)
plt.grid(True, axis='y', linestyle='--', alpha=0.7) # paramentro: linestyle è un indicatore, in questo caso sull'asse delle y
plt.tight_layout()
plt.show()
```

### Analisi tasso di mortalità per continente:
Per ogni continente:

* Calcolare il tasso di mortalità
* visualizzare in un grafico il risultato

```python
# Calcolo del tasso di mortalità per continente
df_latest['mortality_rate'] = (df_latest['total_deaths'] / df_latest['total_cases']) * 100

# Visualizzazione del tasso di mortalità con un grafico a barre orizzontale

plt.figure(figsize=(10, 6))
plt.barh(df_latest.index, df_latest['mortality_rate'], color='darkred')
plt.title('Tasso di Mortalità (%) per Continente')
plt.xlabel('Tasso di Mortalità (%)')
plt.ylabel('Continente')
plt.grid(axis='x', linestyle='dotted', alpha=0.7)  # settato il paramentro linestyle con i puntini piuttosto che i trattini
plt.tight_layout()
plt.show()
```

### Analisi sulla percentuale di test: 
Per ogni continente:
* Rappresenta la distribuzione (in percentuale) dei test totali eseguiti, evidenziando il dato dell'Europa.

```python

colors= ['grey', 'brown', 'orange', 'green', 'blue', 'pink'] #assegno al parametro: colors, una lista di colori
explode=[ 0, 0, 0.3, 0, 0, 0] 
plt.figure(figsize=(8, 8))
plt.pie(df_latest['total_tests'], labels=df_latest.index, autopct='%1.1f%%', startangle=70, colors=colors, explode=explode,
        wedgeprops={'edgecolor':'#000000'}) 
# parametro: autopct, percentuale.
# parametro: startangle, cambia l'angolazione delle fette del grafico
# assegnamo al parametro: explode, una variabile con una lista di valori in oridne posizionale che ne mettono im rilievo la fetta desiderata
# parametro: wedgeprops, colore bordi con esadecimale
plt.title('Distribuzione dei Test Totali per Continente')
plt.tight_layout()
plt.show()

# I dati dell'analisi in numeri
print("\nNumero totale di test per continente:")
print(df_latest['total_tests'].sort_values(ascending=False))
```

### Analisi rapporto vaccinazioni e densità di popolazione:
Per ogni continente:
* Visualizza un grafico con barre sovrapposte che ne evidenzi il rapporto
* Visualizza un grafico di barre affiancate

```python
# Calcolo del numero totale di persone completamente vaccinate per continente
vaccinated_people = df_latest['people_fully_vaccinated']

# Ottenere la densità di popolazione per continente
density = df_latest['population_density']

#Grafico a barre sovrapposte: vaccinazioni e densità di popolazione
plt.figure(figsize=(10, 6))
bar_width = 0.4
continents = df_latest.index
x = range(len(continents))
plt.bar(x, vaccination_rate, width=bar_width, label='Vaccinazioni (%)', color='blue')
plt.bar(x, density, width=bar_width, bottom=vaccination_rate, label='Densità di Popolazione', color='green')
plt.xticks(ticks=x, labels=continents, rotation=45)
plt.title('Vaccinazioni e Densità di Popolazione per Continente (Barre Sovrapposte)')
plt.xlabel('Continente')
plt.ylabel('Valori')
plt.legend()
plt.tight_layout()
plt.show()

# Grafico a barre affiancate: vaccinazioni e densità di popolazione
plt.figure(figsize=(10, 6))
plt.bar([i - bar_width/2 for i in x], vaccination_rate, width=bar_width, label='Vaccinazioni (%)', color='blue') 
# parametro: ba_width/2 indico la visualizzazione in due barre affiancate, altrimenti le vedrei sovrapposte
plt.bar([i + bar_width/2 for i in x], density, width=bar_width, label='Densità di Popolazione', color='green')
plt.xticks(ticks=x, labels=continents, rotation=45)
plt.title('Vaccinazioni e Densità di Popolazione per Continente (Barre Affiancate)')
plt.xlabel('Continente')
plt.ylabel('Valori')
plt.legend()
plt.tight_layout()
plt.show()
```

### Analisi per fasce di età
Per ogni continente:
* Calcolare la percentuale media di persone per fasce di età 65 e 70 
* confrontarle in due grafici affiancati

```python
print( plt.style.available) #stili disponibili per i grafici
```

```python
# Calcolo la media di persone per fasce d'età 
age_data = df.groupby('continent').agg(
    aged_65_older_avg=('aged_65_older', 'mean'),
    aged_70_older_avg=('aged_70_older', 'mean')
).reset_index()

# Visualizzo le percentuali medie per le fasce d'età come con barre orizzontali, utilizzando subpolt 
# che permette di affiancare più grafici nella stessa fig

plt.style.use('ggplot')   # settato uno stile scelto dopo la ricerca fatta nella cella precendente
fig, axes = plt.subplots(1, 2, figsize=(16, 6))
age_data.plot(x='continent', y='aged_65_older_avg', kind='barh', ax=axes[0], color='goldenrod', legend=False)
axes[0].set_title('Percentuale Media di Persone Over 65 per Continente')
axes[0].set_xlabel('Percentuale (%)')
axes[0].set_ylabel('Continente')
axes[0].grid(axis='x', alpha=0.7) # parametro : alpha, rappresenta l'opacitò che è direttamente proporzionale con il valore che gli assegniamo

age_data.plot(x='continent', y='aged_70_older_avg', kind='barh', ax=axes[1], color='darkorange', legend=False)
axes[1].set_title('Percentuale Media di Persone Over 70 per Continente')
axes[1].set_xlabel('Percentuale (%)')
axes[1].set_ylabel('Continente')
axes[1].grid(axis='x', linestyle='--', alpha=0.7) 
# notiamo che il paramentro: linestyle, per alcuni stile, non è possibile settarlo,
# anzi in alcuni casi, elimina del tutto le righe del parametro dal grafico.

plt.tight_layout()
plt.show()
```
